```python
# Rcon untuk key expansion
RCON = [0x01, 0x02, 0x04, 0x08, 0x10, 0x20, 0x40, 0x80, 0x1b, 0x36]


def key_expansion(key):
    """Menghasilkan round keys dari kunci utama"""
    key_words = [key[i:i + 4] for i in range(0, 16, 4)]
    expanded_keys = []

    for i in range(44):
        if i < 4:
            expanded_keys.append(key_words[i])
        else:
            temp = expanded_keys[i - 1][:]
            if i % 4 == 0:
                # Rotasi
                temp = temp[1:] + temp[:1]
                # SubBytes
                for j in range(4):
                    temp[j] = s_box[temp[j]]
                # XOR dengan Rcon
                temp[0] ^= RCON[i // 4 - 1]

            new_word = []
            for j in range(4):
                new_word.append(expanded_keys[i - 4][j] ^ temp[j])
            expanded_keys.append(new_word)

    return expanded_keys
```

```python


def sub_bytes(state):
    """Mengganti setiap byte dengan nilai dari S-box"""
    for i in range(4):
        for j in range(4):
            state[i][j] = s_box[state[i][j]]
    return state


def shift_rows(state):
    """Melakukan pergeseran circular pada baris"""
    state[1] = state[1][1:] + state[1][:1]
    state[2] = state[2][2:] + state[2][:2]
    state[3] = state[3][3:] + state[3][:3]
    return state


def add_round_key(state, round_key):
    """Melakukan XOR state dengan round key"""
    for i in range(4):
        for j in range(4):
            state[i][j] ^= round_key[i][j]
    return state


def mix_columns(state):
    """Mencampur kolom menggunakan transformasi linear"""

    def multiply_by_2(v):
        s = v << 1
        s &= 0xff
        if (v & 128) != 0:
            s = s ^ 0x1b
        return s

    def multiply_by_3(v):
        return multiply_by_2(v) ^ v

    for i in range(4):
        col = [state[j][i] for j in range(4)]

        new_col = [
            multiply_by_2(col[0]) ^ multiply_by_3(col[1]) ^ col[2] ^ col[3],
            col[0] ^ multiply_by_2(col[1]) ^ multiply_by_3(col[2]) ^ col[3],
            col[0] ^ col[1] ^ multiply_by_2(col[2]) ^ multiply_by_3(col[3]),
            multiply_by_3(col[0]) ^ col[1] ^ col[2] ^ multiply_by_2(col[3])
        ]

        for j in range(4):
            state[j][i] = new_col[j]

    return state


```

```python

def inv_sub_bytes(state):
    """Mengganti setiap byte dengan nilai dari Inverse S-box"""
    for i in range(4):
        for j in range(4):
            state[i][j] = inv_s_box[state[i][j]]
    return state


def inv_shift_rows(state):
    """Melakukan pergeseran circular kebalikan pada baris"""
    state[1] = state[1][-1:] + state[1][:-1]
    state[2] = state[2][-2:] + state[2][:-2]
    state[3] = state[3][-3:] + state[3][:-3]
    return state


def inv_mix_columns(state):
    def multiply_by_2(v):
        s = v << 1
        s &= 0xff
        if (v & 128) != 0:
            s = s ^ 0x1b
        return s

    """Mencampur kolom menggunakan transformasi linear invers"""

    def multiply_by_9(v):
        # 9 = 8 + 1
        tmp = multiply_by_2(multiply_by_2(multiply_by_2(v)))
        return tmp ^ v

    def multiply_by_11(v):
        # 11 = 8 + 2 + 1
        tmp = multiply_by_2(multiply_by_2(multiply_by_2(v)))
        return tmp ^ multiply_by_2(v) ^ v

    def multiply_by_13(v):
        # 13 = 8 + 4 + 1
        tmp = multiply_by_2(multiply_by_2(multiply_by_2(v)))
        return tmp ^ multiply_by_2(multiply_by_2(v)) ^ v

    def multiply_by_14(v):
        # 14 = 8 + 4 + 2
        tmp = multiply_by_2(multiply_by_2(multiply_by_2(v)))
        return tmp ^ multiply_by_2(multiply_by_2(v)) ^ multiply_by_2(v)

    for i in range(4):
        col = [state[j][i] for j in range(4)]

        new_col = [
            multiply_by_14(col[0]) ^ multiply_by_11(col[1]) ^ multiply_by_13(col[2]) ^ multiply_by_9(col[3]),
            multiply_by_9(col[0]) ^ multiply_by_14(col[1]) ^ multiply_by_11(col[2]) ^ multiply_by_13(col[3]),
            multiply_by_13(col[0]) ^ multiply_by_9(col[1]) ^ multiply_by_14(col[2]) ^ multiply_by_11(col[3]),
            multiply_by_11(col[0]) ^ multiply_by_13(col[1]) ^ multiply_by_9(col[2]) ^ multiply_by_14(col[3])
        ]

        for j in range(4):
            state[j][i] = new_col[j]

    return state
```

```python

def encrypt(plaintext, key):
    """Fungsi utama untuk enkripsi"""
    state = [[0 for _ in range(4)] for _ in range(4)]

    # Konversi plaintext ke state array
    for i in range(4):
        for j in range(4):
            state[i][j] = plaintext[i + 4 * j]

    # Key expansion
    round_keys = key_expansion(key)

    # Initial round
    add_round_key(state, round_keys[:4])

    # Main rounds
    for round in range(1, 10):
        sub_bytes(state)
        shift_rows(state)
        mix_columns(state)
        add_round_key(state, round_keys[4 * round:4 * (round + 1)])

    # Final round
    sub_bytes(state)
    shift_rows(state)
    add_round_key(state, round_keys[40:])

    # Konversi state array ke ciphertext
    ciphertext = []
    for j in range(4):
        for i in range(4):
            ciphertext.append(state[i][j])

    return bytes(ciphertext)


def decrypt(ciphertext, key):
    """Fungsi utama untuk dekripsi"""
    state = [[0 for _ in range(4)] for _ in range(4)]

    # Konversi ciphertext ke state array
    for i in range(4):
        for j in range(4):
            state[i][j] = ciphertext[i + 4 * j]

    # Key expansion
    round_keys = key_expansion(key)

    # Initial round
    add_round_key(state, round_keys[40:])

    # Main rounds (kebalikan dari enkripsi)
    for round in range(9, 0, -1):
        inv_shift_rows(state)
        inv_sub_bytes(state)
        add_round_key(state, round_keys[4 * round:4 * (round + 1)])
        inv_mix_columns(state)

    # Final round
    inv_shift_rows(state)
    inv_sub_bytes(state)
    add_round_key(state, round_keys[:4])

    # Konversi state array ke plaintext
    plaintext = []
    for j in range(4):
        for i in range(4):
            plaintext.append(state[i][j])

    return bytes(plaintext)
```

```python
import os


# Function to encrypt the image
def encrypt_image(input_path, output_path, key, iv):
    # Read the image file as binary data
    with open(input_path, 'rb') as f:
        image_data = f.read()

    encrypted_data = encrypt(image_data, key)

    # Save the encrypted data to a binary file
    with open(output_path, 'wb') as f:
        f.write(iv + encrypted_data)

    print(f"Image encrypted and saved as {output_path}")


# Function to decrypt the image
def decrypt_image(encrypted_path, output_path, key):
    # Read the encrypted file
    with open(encrypted_path, 'rb') as f:
        file_data = f.read()

    # Extract the IV from the file
    iv = file_data[:16]
    encrypted_data = file_data[16:]

    decrypted_data = decrypt(encrypted_data, key)

    # Save the decrypted data as an image
    with open(output_path, 'wb') as f:
        f.write(decrypted_data)

    print(f"Image decrypted and saved as {output_path}")


```
